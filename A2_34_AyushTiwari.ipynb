```python
# Harris Corner Detection
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('/content/chess.jpeg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)

# Detect corners using Harris
dst = cv2.cornerHarris(gray, blockSize=2, ksize=3, k=0.04)


dst = cv2.dilate(dst, None)


img[dst > 0.01 * dst.max()] = [0, 0, 255]

# Show image
plt.figure(figsize=(8, 6))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Harris Corner Detection')
plt.axis('off')
plt.show()
```

```python
# Shi-Tomasi Corner Detection
import cv2
import numpy as np
import matplotlib.pyplot as plt


img = cv2.imread('/content/chess.jpeg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect corners using Shi-Tomasi
corners = cv2.goodFeaturesToTrack(gray, maxCorners=100, qualityLevel=0.01, minDistance=10)
corners = corners.astype(int)
for corner in corners:
    x, y = corner.ravel()
    cv2.circle(img, (x, y), 4, (0, 255, 0), -1)

plt.figure(figsize=(8, 6))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Shi-Tomasi Corner Detection')
plt.axis('off')
plt.show()
```

```python
# Blob Detection
import cv2
import numpy as np
import matplotlib.pyplot as plt


image_path = input("/content/photo-1460472178825-e5240623afd5.jpeg")

img = cv2.imread(image_path, 0)

detector = cv2.SimpleBlobDetector_create()


keypoints = detector.detect(img)

img_with_blobs = cv2.drawKeypoints(
    img, keypoints, None, (0, 0, 255),
    cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS
)

img_rgb = cv2.cvtColor(img_with_blobs, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.title("Blob Detection")
plt.axis("off")
plt.show()
```
